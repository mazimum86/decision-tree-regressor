## Decision Tree Regression

**Importing the libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Importing the dataset**

```python
data = pd.read_csv('position_salaries.csv')
data.head()
```

```python
X= data.iloc[:,1:-1].values
y=data.iloc[:,-1].values
```

**Train the Decision Tree Regression Model on the whole dataset**

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X,y)
```

**Predicting a new result**

```python
y_pred =regressor.predict([[7.5]])
y_pred
```

**Visualising the Decision Tree Regression Model results (for higher resolution and smoother curve)**

```python
X_ = np.linspace(X.min(),X.max(),1000).reshape(-1,1)
y_pred= regressor.predict(X_)

plt.scatter(X,y,c='red')
plt.plot(X_,y_pred)
plt.title('Truth or Bluff (Decision Tree)')
plt.xlabel('Experience')
plt.ylabel('Salary');
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(15,6))
plot_tree(decision_tree=regressor,fontsize=8);
```
